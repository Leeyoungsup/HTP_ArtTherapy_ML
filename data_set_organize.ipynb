```python
from numpy import copy


def np_to_dict(category_data,npy):
    category_data['file_path'].append(npy[0])
    category_data['x1'].append(npy[1])
    category_data['y1'].append(npy[2])
    category_data['x2'].append(npy[3])
    category_data['y2'].append(npy[4])
    category_data['class_name'].append(npy[5])
    return category_data

def augmentation_data(img,gt):
    
 
    input_img = img[np.newaxis, :, :, :]
    bbox = gt
    
    seq = iaa.Sequential([
        iaa.Fliplr(0.5), # horizontal flips
        iaa.Multiply((0.8, 1.2), per_channel=0.5),
        # Small gaussian blur with random sigma between 0 and 0.5.
        # But we only blur about 50% of all images.
        # Strengthen or weaken the contrast in each image.
        # Add gaussian noise..
        # For 50% of all images, we sample the noise once per pixel.
        # For the other 50% of all images, we sample the noise per pixel AND
        # channel. This can change the color (not only brightness) of the
        # pixels.
        iaa.AdditiveGaussianNoise(loc=0, scale=(0.0, 0.05*255), per_channel=0.5),
        # Make some images brighter and some darker.
        # In 20% of all cases, we sample the multiplier once per channel,
        # which can end up changing the color of the images.
        iaa.Affine(
        scale={"x": (0.5, 0.5), "y": (0.5, 0.5)},
        rotate=(-25, 25),
    )
        
    ], random_order=True)
    
    aug_img, aug_bbox = seq(images = input_img, bounding_boxes = bbox)
    return aug_img[0],aug_bbox

def Annotations(size,count,label_data,file_path,category,file_index=0):
    if(category=='house'):
       global house_Annotations_allcategory_data,ratio_width,ratio_height,X_H,Y_H,house_maincategory_object,house_object
       allcategory_data=house_Annotations_allcategory_data
       maincategory_object=house_maincategory_object
       temp_object=house_object
    elif(category=='tree'):
       global tree_Annotations_allcategory_data,ratio_width,ratio_height,X_H,Y_H,tree_maincategory_object,tree_object
       allcategory_data=tree_Annotations_allcategory_data
       maincategory_object=tree_maincategory_object
       temp_object=tree_object
    elif(category=='person'):
       global person_Annotations_allcategory_data,ratio_width,ratio_height,X_H,Y_H,person_maincategory_object,person_object
       allcategory_data=person_Annotations_allcategory_data
       maincategory_object=person_maincategory_object
       temp_object=person_object
    class_count=len(label_data['objects'])
    img=cv2.imread(file_path)
    img=Image.fromarray(img)
    img1=img.resize(size)
    img_width=img.size[0]
    img_height=img.size[1]
    reduplication_count=0
    gt=[]
    option_list=[]
    class_name1=''
    for i in range(class_count):
        class_name=label_data['objects'][i]['class_name']
        file_meta_index=label_data['objects'][i]['frames'][0]['num']
        if(file_meta_index!=file_index):
            continue
        option_name=''
        for j in range(len(label_data['objects'][i]['properties'])):
            try:
                if(option_name==''):
                    option_name+=label_data['objects'][i]['properties'][j]['option_name']
                else:
                    option_name+=','
                    option_name+=label_data['objects'][i]['properties'][j]['option_name']
            except:
                try:
                    if(option_name==''):
                        option_name+=label_data['objects'][i]['properties'][j]['option_names'][0]
                    else:
                        option_name+=','
                        option_name+=label_data['objects'][i]['properties'][j]['option_names'][0]
                except:
                    option_name=''
        if class_name in maincategory_object.keys():
            class_name1=class_name
            class_name1=temp_object[class_name]
            reduplication_count+=1
        if class_name in temp_object.keys():
            class_name=temp_object[class_name]
            box_data=label_data['objects'][i]['frames'][0]['annotation']['coord']
            ratio_width=size[0]/img_width
            ratio_height=size[1]/img_height
            x=box_data['x']
            y=box_data['y']
            width=box_data['width']
            height=box_data['height']
            x4=(x+width)*ratio_width
            y4=(y+height)*ratio_height
            x3=x*ratio_width
            y3=y*ratio_height
            X_H,Y_H=x3,y3
            option_list.append(option_name)
            gt1=ia.BoundingBox(x1 = x3, y1 = y3, x2 = x4, y2 = y4,label=class_name)
            gt.append(gt1)
    for imgaug_count in range(5):
        img1=np.array(img1)
        gt = BoundingBoxesOnImage(gt, shape=size)
        img_dst,gt_dst=augmentation_data(img1,gt)
        img_dst=Image.fromarray(img_dst)
        if(file_index==2 or file_index==3):
            img_dst.save('./data/'+category+'/origin/'+category+'_'+str(count)+'_'+str(file_index)+'_'+str(imgaug_count)+'.png')
        else:
            img_dst.save('./data/'+category+'/origin/'+category+'_'+str(count)+'_'+str(imgaug_count)+'.png')
        for k in range(len(gt_dst)):
            if(file_index==2 or file_index==3):
                allcategory_data['file_path'].append(category+'/origin/'+category+'_'+str(count)+'_'+str(file_index)+'_'+str(imgaug_count)+'.png')
            else:
                allcategory_data['file_path'].append(category+'/origin/'+category+'_'+str(count)+'_'+str(imgaug_count)+'.png')
            allcategory_data['x1'].append(int(gt_dst[k].x1))
            allcategory_data['y1'].append(int(gt_dst[k].y1))
            allcategory_data['x2'].append(int(gt_dst[k].x2))
            allcategory_data['y2'].append(int(gt_dst[k].y2))
            allcategory_data['class_name'].append(gt_dst[k].label)
            allcategory_data['option_name'].append(option_list[k])
    if(category=='house'):
        house_Annotations_allcategory_data=allcategory_data
        house_maincategory_object=maincategory_object
        house_object=temp_object
    elif(category=='tree'):
        tree_Annotations_allcategory_data=allcategory_data
        tree_maincategory_object=maincategory_object
        tree_object=temp_object
    elif(category=='person'):
        person_Annotations_allcategory_data=allcategory_data
        person_maincategory_object=maincategory_object
        person_object=temp_object
def SubAnnotations(size,main_class,sub_class,allcategory_data,category_data,path):
    numpy_allcatrgory_data=pd.DataFrame(allcategory_data).to_numpy()
    path_list=np.unique(numpy_allcatrgory_data[:,0])
    for temp_path in path_list:
        img=cv2.imread('./data/'+temp_path)
        data_index=np.where(numpy_allcatrgory_data[:,0]==temp_path)[0]
        main_index=np.where(numpy_allcatrgory_data[data_index,5]==main_class)[0]
        if main_index.shape[0] != 0:
            main_data=numpy_allcatrgory_data[data_index[main_index]][0]
            img1=np.copy(img)
            img1[:main_data[2]]=0
            img1[:,:main_data[1]]=0
            img1[main_data[4]:]=0
            img1[:,main_data[3]:]=0
            temp_file_name=temp_path[temp_path.rfind('/'):]
            cv2.imwrite('./data/'+path+temp_file_name,img1)
            for i in data_index:
                if numpy_allcatrgory_data[i,5] in sub_class:
                    x1=numpy_allcatrgory_data[i,1]
                    y1=numpy_allcatrgory_data[i,2]
                    x2=numpy_allcatrgory_data[i,3]
                    y2=numpy_allcatrgory_data[i,4]
                    if x1<0:
                        x1=0
                    if y1<0:
                        y1=0
                    if x2>size[0]:
                        x2=size[0]
                    if y2>size[1]:
                        y2=size[1]
                    category_data['file_path'].append(path+temp_file_name)
                    category_data['x1'].append(int(x1))
                    category_data['y1'].append(int(y1))
                    category_data['x2'].append(int(x2))
                    category_data['y2'].append(int(y2))
                    category_data['class_name'].append(numpy_allcatrgory_data[i,5])
    return category_data
def CropAnnotations(size,main_class,crop_class,sub_class,location,allcategory_data,category_data,path):
    numpy_allcatrgory_data=pd.DataFrame(allcategory_data).to_numpy()
    path_list=np.unique(numpy_allcatrgory_data[:,0])
    for temp_path in path_list:
        img=cv2.imread('./data/'+temp_path)
        data_index=np.where(numpy_allcatrgory_data[:,0]==temp_path)[0]
        main_index=np.where(numpy_allcatrgory_data[data_index,5]==main_class)[0]
        main_data=numpy_allcatrgory_data[data_index[main_index]][0]
        img1=np.copy(img)
        img1[:main_data[2]]=0
        img1[:,:main_data[1]]=0
        img1[main_data[4]:]=0
        img1[:,main_data[3]:]=0
        temp_file_name=temp_path[temp_path.rfind('/'):]
        x3,x4,y3,y4=[],[],[],[]
        for i in data_index:
            if numpy_allcatrgory_data[i,5] in crop_class:
                x1=numpy_allcatrgory_data[i,1]
                y1=numpy_allcatrgory_data[i,2]
                x2=numpy_allcatrgory_data[i,3]
                y2=numpy_allcatrgory_data[i,4]
                if x1<0:
                    x1=0
                if y1<0:
                    y1=0
                if x2>size[0]:
                    x2=size[0]
                if y2>size[1]:
                    y2=size[1]
            if numpy_allcatrgory_data[i,5] in sub_class:
                xx1=numpy_allcatrgory_data[i,1]
                yy1=numpy_allcatrgory_data[i,2]
                xx2=numpy_allcatrgory_data[i,3]
                yy2=numpy_allcatrgory_data[i,4]
                if xx1<0:
                    xx1=0
                if yy1<0:
                    yy1=0
                if xx2>size[0]:
                    xx2=size[0]
                if yy2>size[1]:
                    yy2=size[1]
                x3.append(xx1)
                y3.append(yy1)
                x4.append(xx2)
                y4.append(yy2)
        if len(x3)==0:
            continue
        if location==2:
            img2=np.copy(img1)
            img2[:int(3*y2/4)]=0
        elif location==1:
            img2=np.copy(img1)
            img2[:int(y1+(y2-y1)/2)]=0
        else: 
            img2=np.copy(img1)
            img2[int(y2):]=0    
        cv2.imwrite('./data/'+path+temp_file_name,img2)
        inda=0
        for i in data_index:
            if numpy_allcatrgory_data[i,5] in sub_class:
                
                x1=x3[inda]
                y1=y3[inda]
                x2=x4[inda]
                y2=y4[inda]
                inda+=1
                if x1<0:
                    x1=0
                if y1<0:
                    y1=0
                if x2>size[0]:
                    x2=size[0]
                if y2>size[1]:
                    y2=size[1]
                category_data['file_path'].append(path+temp_file_name)
                category_data['x1'].append(int(x1))
                category_data['y1'].append(int(y1))
                category_data['x2'].append(int(x2))
                category_data['y2'].append(int(y2))
                category_data['class_name'].append(numpy_allcatrgory_data[i,5])
    return category_data
    
def OriginAnnotations(origin_class,allcategory_data,category_data):
    numpy_allcatrgory_data=pd.DataFrame(allcategory_data).to_numpy()
    path_list=np.unique(numpy_allcatrgory_data[:,0])
    for temp_path in path_list:
        data_index=np.where(numpy_allcatrgory_data[:,0]==temp_path)[0]
        for i in data_index:
            if numpy_allcatrgory_data[i,5] in origin_class:
                category_data['file_path'].append(numpy_allcatrgory_data[i,0])
                category_data['x1'].append(numpy_allcatrgory_data[i,1])
                category_data['y1'].append(numpy_allcatrgory_data[i,2])
                category_data['x2'].append(numpy_allcatrgory_data[i,3])
                category_data['y2'].append(numpy_allcatrgory_data[i,4])
                category_data['class_name'].append(numpy_allcatrgory_data[i,5])
    return category_data
def tree_pillar_to_annual_zone(allcategory_data):
    numpy_allcatrgory_data=pd.DataFrame(allcategory_data).to_numpy()
    path_list=np.unique(numpy_allcatrgory_data[:,0])
    for temp_path in path_list:
        data_index=np.where(numpy_allcatrgory_data[:,0]==temp_path)[0]
        pillar_index=0
        extent=0
        for i in data_index:
            if numpy_allcatrgory_data[i,5]=='pillar':
                temp_extent=(numpy_allcatrgory_data[i,3]-numpy_allcatrgory_data[i,1])*(numpy_allcatrgory_data[i,4]-numpy_allcatrgory_data[i,2])
                if extent<temp_extent:
                    extent=temp_extent
                    pillar_index=i
                    
        for j in data_index:
            if numpy_allcatrgory_data[j,5]=='pillar':
                if j!=pillar_index:
                    allcategory_data['class_name'][j]='annual_zone'
        
    return allcategory_data
```
